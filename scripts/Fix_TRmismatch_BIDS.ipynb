```python
layout = bids.BIDSLayout("../../../../mnt/c/Users/justi/Documents/Universite_Laval/Doctorat/Cours/Brainhack/BIDS")
layout
```

```python
bold_images = layout.get(suffix="bold", extension=[".nii", ".nii.gz"])
```

```python
for bold in bold_images:
    TR = bold.get_metadata()["RepetitionTime"]
    img = bold.get_image()
    zooms = img.header.get_zooms()
    if not np.isclose(zooms[3], TR):
        old_path = Path(bold)
        new_path = old_path.with_name(f"{old_path.stem}_fixed.nii.gz")
        img.header.set_zooms(zooms[:3] + (TR,))
        img.to_filename(new_path)
```
